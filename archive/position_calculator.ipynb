```python
# Trading Position Calculator - Entry/SL/TP basiert

# Inputs
capital = float(input("Account Size: "))
risk_pct = float(input("Risk %: "))
entry_price = float(input("Entry Price: "))
sl_price = float(input("Stop Loss Price: "))
tp_price = float(input("Take Profit Price: "))

# Erkenne ob Long oder Short
if tp_price > entry_price:
    trade_direction = "LONG"
    is_long = True
elif tp_price < entry_price:
    trade_direction = "SHORT"
    is_long = False
else:
    print("Error: TP muss unterschiedlich zu Entry sein!")
    exit()

# Validiere SL Position
if is_long and sl_price >= entry_price:
    print("‚ö†Ô∏è ERROR: Bei LONG muss SL unter Entry sein!")
    exit()
elif not is_long and sl_price <= entry_price:
    print("‚ö†Ô∏è ERROR: Bei SHORT muss SL √ºber Entry sein!")
    exit()

print(f"\nüéØ Trade Direction: {trade_direction}")

# Berechne SL und TP in Prozent
sl_distance = abs((entry_price - sl_price) / entry_price) * 100
tp_distance = abs((tp_price - entry_price) / entry_price) * 100

# Risk/Reward Ratio
risk_reward = tp_distance / sl_distance

# Position Size Berechnung
risk_amount = capital * (risk_pct / 100)

# Berechne ben√∂tigte Position Size basierend auf Risk
# Position Size = Risk Amount / SL Distance (in decimals)
position_size = risk_amount / (sl_distance / 100)

# Berechne ben√∂tigte Margin und Leverage
if position_size <= capital:
    # Kein Leverage n√∂tig
    margin = position_size
    leverage = 1.0
else:
    # Leverage n√∂tig
    margin = capital
    leverage = position_size / capital

# Potentielle Gewinne/Verluste in $
potential_loss = risk_amount
potential_profit = (tp_distance / sl_distance) * risk_amount

# Output
print(f"\n--- Trade Analysis ---")
print(f"Entry: ${entry_price:.2f}")
print(f"Stop Loss: ${sl_price:.2f} ({sl_distance:.2f}%)")
print(f"Take Profit: ${tp_price:.2f} ({tp_distance:.2f}%)")
print(f"\nRisk/Reward: 1:{risk_reward:.2f}")

print(f"\n--- Position Details ---")
print(f"Margin: ${margin:.2f}")
print(f"Leverage: {leverage:.2f}x")
print(f"Position Size: ${position_size:.2f}")

print(f"\n--- P&L ---")
print(f"Max Loss: ${potential_loss:.2f}")
print(f"Max Profit: ${potential_profit:.2f}")

# Warnings
if risk_reward < 2:
    print("\n‚ö†Ô∏è WARNING: Risk/Reward below 1:2!")
if leverage > 10:
    print("‚ö†Ô∏è WARNING: High leverage detected!")
if leverage > 1:
    print(f"‚ÑπÔ∏è INFO: Leverage needed because position ({position_size:.2f}) > capital ({capital:.2f})")
```

```python
# Trading Position Calculator - Entry/SL/TP basiert

# Inputs
capital = float(input("Account Size: "))
risk_pct = float(input("Risk %: "))
entry_price = float(input("Entry Price: "))
sl_price = float(input("Stop Loss Price: "))
tp1_price = float(input("Take Profit 1 Price: "))
tp2_price = float(input("Take Profit 2 Price: "))

# Erkenne ob Long oder Short (basierend auf TP2)
if tp2_price > entry_price:
    trade_direction = "LONG"
    is_long = True
elif tp2_price < entry_price:
    trade_direction = "SHORT"
    is_long = False
else:
    print("Error: TP muss unterschiedlich zu Entry sein!")
    exit()

# Validiere SL Position
if is_long and sl_price >= entry_price:
    print("‚ö†Ô∏è ERROR: Bei LONG muss SL unter Entry sein!")
    exit()
elif not is_long and sl_price <= entry_price:
    print("‚ö†Ô∏è ERROR: Bei SHORT muss SL √ºber Entry sein!")
    exit()

# Validiere TP1 und TP2 Reihenfolge
if is_long:
    if tp1_price <= entry_price or tp2_price <= entry_price:
        print("‚ö†Ô∏è ERROR: Bei LONG m√ºssen TPs √ºber Entry sein!")
        exit()
    if tp1_price >= tp2_price:
        print("‚ö†Ô∏è ERROR: TP1 muss unter TP2 liegen!")
        exit()
else:
    if tp1_price >= entry_price or tp2_price >= entry_price:
        print("‚ö†Ô∏è ERROR: Bei SHORT m√ºssen TPs unter Entry sein!")
        exit()
    if tp1_price <= tp2_price:
        print("‚ö†Ô∏è ERROR: TP1 muss √ºber TP2 liegen!")
        exit()

print(f"\nüéØ Trade Direction: {trade_direction}")

# Berechne SL und TP in Prozent
sl_distance = abs((entry_price - sl_price) / entry_price) * 100
tp1_distance = abs((tp1_price - entry_price) / entry_price) * 100
tp2_distance = abs((tp2_price - entry_price) / entry_price) * 100

# Risk/Reward Ratios
risk_reward_tp1 = tp1_distance / sl_distance
risk_reward_tp2 = tp2_distance / sl_distance

# Position Size Berechnung
risk_amount = capital * (risk_pct / 100)
position_size = risk_amount / (sl_distance / 100)

# Berechne ben√∂tigte Margin und Leverage
if position_size <= capital:
    margin = position_size
    leverage = 1.0
else:
    margin = capital
    leverage = position_size / capital

# Potentielle Gewinne/Verluste in $ (mit 50/50 Split)
potential_loss = risk_amount

# 50% der Position bei TP1, 50% bei TP2
profit_at_tp1 = (position_size * 0.5) * (tp1_distance / 100)
profit_at_tp2 = (position_size * 0.5) * (tp2_distance / 100)
potential_profit_total = profit_at_tp1 + profit_at_tp2

# Average R/R bei 50/50 Split
avg_risk_reward = (risk_reward_tp1 + risk_reward_tp2) / 2

# Output
print(f"\n--- Trade Analysis ---")
print(f"Entry: ${entry_price:.2f}")
print(f"Stop Loss: ${sl_price:.2f} ({sl_distance:.2f}%)")
print(f"Take Profit 1 (50%): ${tp1_price:.2f} ({tp1_distance:.2f}%) - R/R 1:{risk_reward_tp1:.2f}")
print(f"Take Profit 2 (50%): ${tp2_price:.2f} ({tp2_distance:.2f}%) - R/R 1:{risk_reward_tp2:.2f}")
print(f"\nAverage R/R: 1:{avg_risk_reward:.2f}")

print(f"\n--- Position Details ---")
print(f"Margin: ${margin:.2f}")
print(f"Leverage: {leverage:.2f}x")
print(f"Position Size: ${position_size:.2f}")
print(f"Size at TP1 (50%): ${position_size * 0.5:.2f}")
print(f"Size at TP2 (50%): ${position_size * 0.5:.2f}")

print(f"\n--- P&L ---")
print(f"Max Loss: ${potential_loss:.2f}")
print(f"Profit at TP1: ${profit_at_tp1:.2f}")
print(f"Profit at TP2: ${profit_at_tp2:.2f}")
print(f"Total Profit (if both hit): ${potential_profit_total:.2f}")

# Warnings
if avg_risk_reward < 2:
    print("\n‚ö†Ô∏è WARNING: Average Risk/Reward below 1:2!")
if leverage > 10:
    print("‚ö†Ô∏è WARNING: High leverage detected!")
if leverage > 1:
    print(f"‚ÑπÔ∏è INFO: Leverage needed because position ({position_size:.2f}) > capital ({capital:.2f})")
```
